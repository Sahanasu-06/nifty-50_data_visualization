Changing Object type of Open, High, Low,LTP to float

```python
df['Open'] = df['Open'].str.replace(',','')
df['Open'] = df['Open'].astype(float)
```

```python
df.head(5)
```

```python
df['High'] = df['High'].str.replace(',','')
df['High'] = df['High'].astype(float)
```

```python
df['Low'] = df['Low'].str.replace(',','')
df['Low'] = df['Low'].astype(float)
```

```python
df.head()
```

```python
df['LTP'] = df['LTP'].str.replace(',','')
df['LTP'] = df['LTP'].astype(float)
```

```python
df.dtypes
```

```python
df['Turnover (crs.)'] = df['Turnover (crs.)'].str.replace(',','')
df['Turnover (crs.)'] = df['Turnover (crs.)'].astype(float)
```

```python
df['52w H'] = df['52w H'].str.replace(',','')
```

```python
df['52w H'] = df['52w H'].astype(float)
```

```python
df['52w L'] = df['52w L'].str.replace(',','')
df['52w L'] = df['52w L'].astype(float)
```

```python
df.dtypes
```

Now we have every column in float

```python
df.describe()
```

```python
df['Open'] = df['Open'].astype(int)
df['High'] = df['High'].astype(int)
df['Low'] = df['Low'].astype(int)
df['LTP'] = df['LTP'].astype(int)
df['Chng'] = df['Chng'].astype(int)
df['% Chng'] = df['% Chng'].astype(int)
df['Volume (lacs)'] = df['Volume (lacs)'].astype(int)
df['Turnover (crs.)'] = df['Turnover (crs.)'].astype(int)
df['52w H'] = df['52w H'].astype(int)
df['365 d % chng'] = df['365 d % chng'].astype(int)
df['30 d % chng	'] = df['30 d % chng'].astype(int)
```

```python
df.head()
```

Creating group of Banking and Financial Companies in this dataset.

```python
bank_df = df['Symbol'].isin(['AXISBANK','HDFCBANK','ICICIBANK','INDUSINDBK','KOTAKBANK','SBIN','HDFC','BAJAJFINSV','BAJFINANCE','HDFCLIFE'])
bank_table = df[bank_df].pivot_table(values =['Open','High','Low','LTP','Chng','Volume (lacs)','52w H','52w L','Turnover (crs.)','365 d % chng','30 d % chng'], index= ['Symbol'])
bank_table.reset_index(level=0, inplace=True)
bank_table
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=45)


sns.barplot(data = bank_table, x = 'Symbol', y = '365 d % chng', palette='flare')


plt.xlabel("Symbol")  # add X-axis label
plt.ylabel('% Change in a Year')  # add Y-axis label
plt.title("Yearly Growth")  # add title


plt.show
```

As the above data, BajajFinserv and SBIN have shown extremely positive growth in comparison to its peers. HDFC LIFE has shown no growth against its peers. Same applies for majority of banks.

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=45)


sns.barplot(data = bank_table, x = 'Symbol', y = '30 d % chng',palette='mako')
```

AxisBank and IndusInd Bank have been the worst performers with more than 20% wealth loss of the investors in a single month

```python
sns.pairplot(bank_table,hue = 'Symbol',x_vars = ('Volume (lacs)',"Turnover (crs.)","30 d % chng",'365 d % chng'),
             y_vars=('52w H','52w L'),height=3,palette='icefire')
```

The above pairplot showcases that Axis Bank with highest turnover doesnot seem to have any correlation with its peers while rest of stock seems to be following a range bound pattern.
